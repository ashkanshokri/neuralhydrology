```python
import pandas as pd
from pathlib import Path
from matplotlib import pyplot as plt
import numpy as np

```

```python
variable = 'NSE'
periods = ['test', 'train']
runs = ['train_2000_to_2018_2201_092259', 'r_pet_train_2000_to_2018_2201_220657']

legend = []
for run in runs:
    for period in periods:
        p = Path(f'runs/{run}/{period}/model_epoch030/{period}_metrics.csv')
        legend.append(p)
        df = pd.read_csv(p, index_col=0)
        y = df[variable].sort_values(ascending=False).values
        x = np.linspace(0, 1, len(y))
        plt.plot(x, y)

plt.ylim([-1, 1])
plt.xlim([0, 1])
plt.legend(legend)
plt.ylabel(variable)
plt.xlabel('Probability of exceedance')
```
